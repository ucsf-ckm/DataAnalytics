```python
import pandas as pd

from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())
```

```python
subjects = pd.read_csv("subject.csv")
```

```python
subjects
```

```python
measurements = pd.read_csv("measurement.csv")
```

```python
measurements
```

```python
combined_df = pysqldf("SELECT * FROM subjects JOIN measurements \
    ON subjects.subject_id = measurements.subject_id \
    WHERE gender = 'M' AND data > 105")
```
