```python
# extract the staff names from the profiles directory in the format firstname.lastname
```

```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://profiles.ucsf.edu/sitemap.xml"
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, "lxml")
```

```python
names = soup.find_all('loc')
```

```python
#print(names)
```

```python
staff_names = []
for n in names:
    print(str(n).split('/')[3][:-1])
    staff_names.append(str(n).split('/')[3][:-1])
```

```python
for i in range(10, len(staff_names)):
    print(staff_names[i])
```
